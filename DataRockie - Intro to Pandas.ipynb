```python
import pandas as pd
```

```python
# read csv file
penguins = pd.read_csv("penguins.csv")
```

```python
# head
penguins.head()
```

```python
# tail 
penguins.tail()
```

```python
# see dataframe dimension
penguins.shape
```

```python
# penguins info
penguins.info()
```

```python
# summary statistics
penguins.describe()
```

```python
# remove rows with missing values
clean_penguins = penguins.dropna()
clean_penguins.shape
```

```python
# check number of null values in each column
clean_penguins.isnull().sum()
```

```python
# select a single column
clean_penguins['species']
```

```python
# select more than one column
clean_penguins[['species', 'island', 'sex']].head()
```

```python
# rename column sex to gender
clean_penguins = clean_penguins.rename(columns = {'sex': 'gender'})
clean_penguins.head()
```

```python
# drop columns 
small_penguins = clean_penguins.drop(columns = ['gender', 'body_mass_g', 'island'])
small_penguins.head()
```

```python
# iloc slice rows 0:5
small_penguins.iloc[0:5]
```

```python
# loc filter rows with condition
small_penguins.loc[ small_penguins['bill_length_mm'] >= 55 ]
```

```python
# query same result
small_penguins.query( "bill_length_mm >= 55" )
```

```python
# sort values descending order (high to low)
small_penguins.sort_values(by = 'bill_length_mm', ascending = False)
```

```python
# write csv file
small_penguins.to_csv("small_penguins.csv")
```
